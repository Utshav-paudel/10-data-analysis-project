## Data cleaning

#### 1. Instruction ( For Data Cleaning ) - Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
data.head(2)
```

```python
column = ['EngineSize','Cylinders','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']
for col in column:
    data[col].fillna(data[col].mean(),inplace = True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```
