### 3.Does gender affect who gets searched during a stop ?
Question ( mapping + data-type casting )

```python
df.head()
```

```python
df.groupby('driver_gender').search_conducted.sum()
```

```python
df.search_conducted.value_counts()  #checking how many times search was conducted
```

#### Search was conducted 2479 times out of which male were searched 2113 times.

So, male are searched more often

### 4.What is the mean stop_duration ?

```python
df.head()
```

```python
df.stop_duration.dtype
```

```python
df.stop_duration.value_counts()
```

```python
df['stop_duration'] = df['stop_duration'].map({'0-15 Min' : 7.5, '16-30 Min' : 23,'30+ Min' : 30 })
```

```python
df.head()
```

```python
df.stop_duration.mean()
```

### The mean stop_duration is around 11 minute.
